**Preprocessing Of Dataset**

---



```python
n_class = 3
n1 = 1121
n2 = 303
n3 = 1765
```

```python
df_pos = df.copy()[df.label == 0][:n1]
df_neg = df.copy()[df.label == 1][:n2]
df_neu = df.copy()[df.label == 2][:n3]
combi = pd.concat([df_pos, df_neg, df_neu], ignore_index=True).reset_index(drop=True)
```

```python
def remove_pattern(input_txt, pattern):
  r = re.findall(pattern, input_txt)
  for i in r:
    input_txt = re.sub(i, '', input_txt)

  return input_txt
```

```python
combi.head()
```

```python
combi['tweet'] = np.vectorize(remove_pattern)(combi['tweet'], "@[\w]*")

```

```python
combi.head()
```

```python
combi['tweet'] = combi['tweet'].str.replace("[^a-zA-Z#]", " " )
```

```python
combi.head()
```

```python
combi['tweet'] = combi['tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
```

```python
combi.head()
```

**Tokenization**

---



```python
tokenized_tweet = combi['tweet'].apply(lambda x: x.split())
```

**Stemming**

---



```python
from nltk.stem.porter import *
stemmer = PorterStemmer()
```

```python
tokenized_tweet = tokenized_tweet.apply(lambda x: [stemmer.stem(i) for i in x])
```

```python
for i in range(len(tokenized_tweet)):
  tokenized_tweet[i] = ' '.join(tokenized_tweet[i])

combi['tweet'] = tokenized_tweet
```

**Dealing with hashtags**


---



```python
def hashtag_extract(x):
  hashtags = []
  for i in x:
    ht = re.findall(r"#(\w+)", i)
    hashtags.append(ht)

  return hashtags
```

```python
HT_regular = hashtag_extract(combi['tweet'][combi['label'] == 0])

HT_negative = hashtag_extract(combi['tweet'][combi['label'] == 1])

HT_hateful = hashtag_extract(combi['tweet'][combi['label'] == 2])

HT_regular = sum(HT_regular, [])
HT_negative = sum(HT_negative, [])
HT_hateful = sum(HT_hateful, [])
```

2. Using Naive Bayes Classifier

---





```python
from sklearn.model_selection import train_test_split
from collections import Counter
```

```python
X_train, X_test, y_train, y_test = train_test_split(combi['tweet'], combi['label'], test_size=0.33, random_state=0)
```

```python
df_train = pd.DataFrame()
df_test = pd.DataFrame()

df_train['tweet'] = X_train
df_train['label'] = y_train
df_train = df_train.reset_index(drop=True)

df_test['tweet'] = X_test
df_test['label'] = y_test
df_test = df_test.reset_index(drop=True)
```

```python
class TweetNBClassifier(object):

    def __init__(self, df_train):
        self.df_train = df_train
        self.df_pos = df_train.copy()[df_train.label == 0]
        self.df_neg = df_train.copy()[df_train.label == 1]
        self.df_neu = df_train.copy()[df_train.label == 2]

    def fit(self):
        Pr_pos = df_pos.shape[0]/(self.df_train.shape[0])
        Pr_neg = df_neg.shape[0]/(self.df_train.shape[0])
        Pr_neu = df_neu.shape[0]/(self.df_train.shape[0])
        self.Prior  = (Pr_pos, Pr_neg, Pr_neu)

        self.pos_words = ' '.join(self.df_pos['tweet'].tolist()).split()
        self.neg_words = ' '.join(self.df_neg['tweet'].tolist()).split()
        self.neu_words = ' '.join(self.df_neu['tweet'].tolist()).split()

        all_words = ' '.join(self.df_train['tweet'].tolist()).split()

        self.vocab = len(Counter(all_words))

        wc_pos = len(' '.join(self.df_pos['tweet'].tolist()).split())
        wc_neg = len(' '.join(self.df_neg['tweet'].tolist()).split())
        wc_neu = len(' '.join(self.df_neu['tweet'].tolist()).split())
        self.word_count = (wc_pos, wc_neg, wc_neu)
        return self


    def predict(self, df_test):
        class_choice = [0, 1, 2]

        classification = []
        for tweet in df_test['tweet']:
            text = tweet.split()

            val_pos = np.array([])
            val_neg = np.array([])
            val_neu = np.array([])
            for word in text:
                tmp_pos = np.log((self.pos_words.count(word)+1)/(self.word_count[0]+self.vocab))
                tmp_neg = np.log((self.neg_words.count(word)+1)/(self.word_count[1]+self.vocab))
                tmp_neu = np.log((self.neu_words.count(word)+1)/(self.word_count[2]+self.vocab))
                val_pos = np.append(val_pos, tmp_pos)
                val_neg = np.append(val_neg, tmp_neg)
                val_neu = np.append(val_neu, tmp_neu)

            val_pos = np.log(self.Prior[0]) + np.sum(val_pos)
            val_neg = np.log(self.Prior[1]) + np.sum(val_neg)
            val_neu = np.log(self.Prior[2]) + np.sum(val_neu)

            probability = (val_pos, val_neg, val_neu)
            classification.append(class_choice[np.argmax(probability)])
        return classification


    def score(self, feature, target):

        compare = []
        for i in range(0,len(feature)):
            if feature[i] == target[i]:
                tmp ='correct'
                compare.append(tmp)
            else:
                tmp ='incorrect'
                compare.append(tmp)
        r = Counter(compare)
        accuracy = r['correct']/(r['correct']+r['incorrect'])
        return accuracy
```

```python
tnb = TweetNBClassifier(df_train)
tnb = tnb.fit()
```
